```python
from pop import Potentiometer, delay
poten = Potentiometer()
while True:
    print(f"Potenmeter.read : {poten.read()}")
    print(f"Potenmeter.readAverage : {poten.readAverage()}")
    print(f"Potenmeter.readVolt : {poten.readVolt()}")
    print(f"Potenmeter.readVoltAverage : {poten.readVoltAverage()}")
    print()
    delay(1000)
    
```
