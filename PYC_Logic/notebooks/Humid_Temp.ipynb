```python
from pop import delay, Oled, Sht20, Switches
display = Oled()
sht = Sht20()
btn = Switches()

# 초기화
display.init()
display.setTextSize(s=1)
display.clearDisplay()

while True:
    temp = sht.readTemp()
    humi = sht.readHumi()
    
    display.clearDisplay()
    
    display.setCursor(10, 10)    
    display.print("Temp : {:.2f}'C".format(float(temp)))
    
    display.setCursor(10, 20)
    display.print("humi : {:.2f}%".format(float(humi)))
    
    display.display()
    delay(1000)
```
