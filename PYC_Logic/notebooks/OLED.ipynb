```python
# Oled size = 128, 64
from pop import Oled, delay
oled = Oled() 
oled.init()
(x, y) = (0, 0)# 시작 위치
oled.setTextSize(s=1) # 글자 위치
while True:
    if x > 128 and y > 64:
        x = y = 0 #50을 넘어가면 원상태로 위치
    oled.setCursor(x, y)
    oled.print("kyj")
    oled.clearDisplay()
    (x, y) = (x+1, y+1)
```

```python
from pop import Oled, delay, Switches

display = Oled()
btn = Switches()

forwardReverse = True   # True: 정삼각형, False: 역삼각형
colorOnOff = False      # True: 채움, False: 빈 삼각형

prev0 = True
prev1 = True

def redraw():
    display.clearDisplay()

    if forwardReverse:   # 위를 향하는 삼각형
        x1, y1 = 64, 8
        x2, y2 = 32, 56
        x3, y3 = 96, 56
    else:                # 아래를 향하는 삼각형
        x1, y1 = 32, 8
        x2, y2 = 96, 8
        x3, y3 = 64, 56

    if colorOnOff:
        display.fillTriangle(x1, y1, x2, y2, x3, y3, display.WHITE)
    else:
        display.drawTriangle(x1, y1, x2, y2, x3, y3, display.WHITE)

    display.display()

redraw()

while True:
    b0 = btn[0].read()  # 눌리면 False
    b1 = btn[1].read()  # 눌리면 False

    changed = False

    if prev0 and (not b0):   # 버튼0 눌림 1회
        forwardReverse = not forwardReverse
        changed = True

    if prev1 and (not b1):   # 버튼1 눌림 1회
        colorOnOff = not colorOnOff
        changed = True

    prev0 = b0
    prev1 = b1

    if changed:
        redraw()

    delay(20)
```
