```python
from pop import Sht20, delay
sht = Sht20 # Sht20 온도/습소 센서 사용
with open("Temperature_Humidity.csv", 'w', encoding='utf-8') as f:
    while True:
        temp = sht.readTemp()
        humi = sht.readHumi()
        print(f"온도 :  {temp:.2f}")
        print(f"습도 : {humi:.2f}")
        f.write(f"{temp:.2f}"+", ")
        f.write(f'{humi:.2f}'+'\n')
        delay(2000)
```
