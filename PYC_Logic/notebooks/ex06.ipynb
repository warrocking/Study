```python
from pop import Sht20, delay
sht = Sht20() # Sht20 온도/습소 센서 사용
with open("Temperature_Humidity.csv", 'w', encoding='utf-8') as f:
    while True:
        temp = sht.readTemp()
        humi = sht.readHumi()
        print(f"온도 :  {temp:.2f}")
        print(f"습도 : {humi:.2f}")
        f.write(f"{temp:.2f}"+", ")
        f.write(f'{humi:.2f}'+'\n')
        delay(2000)
```

```python
from pop import Sht20, delay
import csv
from datetime import datetime

sht = Sht20()

with open("Temperature_Humidity.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["time", "temp_C", "humi_percent"])  # 헤더

    try:
        while True:
            temp = sht.readTemp()
            humi = sht.readHumi()
            now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

            print(f"온도: {temp:.2f} C")
            print(f"습도: {humi:.2f} %")

            writer.writerow([now, f"{temp:.2f}", f"{humi:.2f}"])
            f.flush()   # 중간중간 즉시 저장
            delay(2000)

    except KeyboardInterrupt:
        print("저장 종료: Temperature_Humidity.csv")
```

```python
from pop import Sht20, delay
import csv

sht = Sht20()

with open("temp_humi.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["num", "temp_C", "humi_percent"])  # 헤더
    
    try:
        for i in range(0, 100):
            temp = sht.readTemp()
            humi = sht.readHumi()

            print(f"온도: {temp:.2f} C")
            print(f"습도: {humi:.2f} %")
            num = i+1
            writer.writerow([num, f"{temp:.2f}", f"{humi:.2f}"])
            f.flush()   # 중간중간 즉시 저장
            delay(2000)

    except KeyboardInterrupt:
        print("저장 종료: Temperature_Humidity.csv")
```
