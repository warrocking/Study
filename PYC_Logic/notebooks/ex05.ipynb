```python
from pop import delay, Leds, Cds, Switches, Potentiometer, Psd
poten = Potentiometer()
led = Leds()
def ledon(n):
    if n==0:
        led[0].on
    else:
        for i in range(n):
            led[i].on()
    
def ledoff(n):
    for i in range(n):
        led[i].off()
while True:
    poten_value = poten.readVolt()
    print(poten_value)
    if 0.5<poten_value:
        led[0].on()
    if 1<poten_value:
        led[1].on()
    if 1.5<poten_value:
        led[2].on()
    if 2<poten_value:
        led[3].on()
    if 2.5<poten_value:
        led[4].on()
    if 3<poten_value:
        led[5].on()
    if 3.5<poten_value:
        led[6].on()
    delay(1000)
```
