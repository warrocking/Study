```python
from pop import Gesture, delay, Switches
gesture_proximity = Gesture.Proximity()
btn = Switches()
while True:
    btn_1 = btn[0].read()
    for i in range(10):
        distance = gesture_proximity.read()
        print(distance)
        delay(100)
    if not btn_1:
        break
```

```python
# 색상 감지하기
from pop import Gesture, Switches
gesture_color = Gesture.Color()
btn = Switches()
while True:
    btn_0 = btn[0].read()
    for i in range(10):
        (RED, GREEN, BLUE) = (gesture_color.readRed(), gesture_color.readGreen(), gesture_color.readBlue())
        print("RED = {}, GREEN = {}, BLUE = {}".format(RED, GREEN, BLUE))
        delay(100)
    if not btn_0:
        break
```

```python
# left -> 7 led on
# right -> 7 led off
# up -> 1 led on
# down -> 1 led off
# near -> led all on
# far -> led all off

from pop import Gesture, Leds, Switches, delay
led = Leds()
gs = Gesture()
btn = Switches()

while True:
    btn_0 = btn[0].read()
    btn_1 = btn[1].read()
    
    if not btn_0:
        print("Gesture On")
        if gs.isAvailable():
            motion = gs.readStr()
            print(motion)
            if motion == "Up":
                led[1].on()
                print("led 1 on")
            elif motion == "Down":
                led[1].off()
                print("led 1 off")
            elif motion == "Left":
                led[7].on()
                print("led 7 on")
            elif motion == "Right":
                led[7].off()
                print("led 7 off")
            elif motion == "near":
                led.allOff()
                print("led all off")
            elif motion == "far":
                led.allOn()
                print("led 7 on")
            delay(100)
    if not btn_1:
        break   
```
