```python
from pop import delay, Leds, Cds, Switches, Potentiometer, Psd

led = Leds()
cds = Cds()
switch = Switches()
poten = Potentiometer()
psd = Psd()

print('start: s1/s2는 눌리면 False 입니다.')

tick = 0
try:
    while True:
        s1 = switch[0].read()
        s2 = switch[1].read()
        cds_val = cds.read()
        dist = psd.calcDist(psd.readAverage())
        poten_value = poten.readVolt()

        
        print("poten = ", poten_value)
        if poten_value >= 3:
            print(f'종료: poten={poten_value:.2f}')
            break

        # 준비 완료 조건 : 조도 700 아래에 적외선 사이 거리 10cm 밑으로
        ready = (cds_val < 700) and (dist < 10)

        # ready체크 10회에 1초이며 이를 이용해 1초에 한번씩 내용출력.
        if tick % 10 == 0:
            print(f's1={s1} s2={s2} cds={cds_val} dist={dist:.2f} ready={ready}')

        if not ready:
            led.allOff()
            delay(100)
            tick += 1
            continue

        # 첫번째 버튼 눌렀을 때
        if not s1: 
            for i in range(0, 8):
                led[i].on()
                delay(120)
            for i in range(7, -1, -1):
                led[i].off()
                delay(120)
        
        # 두번째 버튼 눌렀을 때
        elif not s2:
            for i in range(7, -1, -1):
                led[i].on()
                delay(120)
            for i in range(0, 8):
                led[i].off()
                delay(120)
        else:
            led.allOff() # read 이외 상황에는 led 다 꺼놓기

        # 이건 AI 추천으로 작성
        # 출력 주기 제어용으로 사용을 위해 동작 카운트와 로그 카운트 분리를 위해서 라는데 이유는 모르겠음.
        delay(50)
        tick += 1

except KeyboardInterrupt:
    print('중지')
finally:
    led.allOff()
```
