```python
from pop import PixelDisplay, delay

pd = PixelDisplay(automode=False)

def pick_color(names, fallback):
    for n in names:
        if hasattr(pd, n):
            return getattr(pd, n)
    return fallback  # 상수명이 다를 때 대비

BLACK = 0
WHITE = pick_color(["WHITE", "White"], 0xFFFFFF)
RED   = pick_color(["RED", "Red"], 0xFF0000)
GREEN = pick_color(["GREEN", "Green"], 0x00FF00)
BLUE  = pick_color(["BLUE", "Blue"], 0x0000FF)

print("size:", pd.width, "x", pd.height)
print("display methods:", [m for m in ["fill", "setColor", "display", "setBrightness"] if hasattr(pd, m)])

if hasattr(pd, "setBrightness"):
    pd.setBrightness(40)  # 너무 어두우면 숫자 올리세요

# 1) 전체 색상 점등 테스트
for c, name in [(RED, "RED"), (GREEN, "GREEN"), (BLUE, "BLUE"), (WHITE, "WHITE")]:
    pd.fill(c)
    pd.display()
    print("fill:", name)
    delay(500)

# 2) 픽셀 스캔 테스트 (좌상단 -> 우하단)
pd.fill(BLACK)
pd.display()
for y in range(pd.height):
    for x in range(pd.width):
        pd.setColor(x, y, WHITE)
        pd.display()
        delay(20)

# 3) 체커보드 테스트
for y in range(pd.height):
    for x in range(pd.width):
        pd.setColor(x, y, RED if (x + y) % 2 == 0 else BLUE)
pd.display()
delay(1000)

# 종료: 끄기
pd.fill(BLACK)
pd.display()
print("PixelDisplay test done")
```
