# PixelDisplay Diagnostics
Run this notebook top to bottom to check:
1) Brightness/display output
2) GPIO permission/device status
3) Audio conflict config (`dtparam=audio`)
4) Non-root vs root terminal behavior

````python
from pop import PixelDisplay, delay

pd = PixelDisplay(automode=False)

def get_color(name_list, fallback):
    for n in name_list:
        if hasattr(pd, n):
            return getattr(pd, n)
    return fallback

BLACK = 0
WHITE = get_color(['WHITE', 'White'], 0xFFFFFF)
RED = get_color(['RED', 'Red'], 0xFF0000)
GREEN = get_color(['GREEN', 'Green'], 0x00FF00)
BLUE = get_color(['BLUE', 'Blue'], 0x0000FF)

print('size:', pd.width, 'x', pd.height)
print('has setBrightness:', hasattr(pd, 'setBrightness'))

if hasattr(pd, 'setBrightness'):
    for b in [10, 30, 80, 160, 255]:
        pd.setBrightness(b)
        pd.fill(WHITE)
        pd.display()
        print(f'brightness={b} -> white fill')
        delay(700)

for c, n in [(RED, 'RED'), (GREEN, 'GREEN'), (BLUE, 'BLUE'), (WHITE, 'WHITE')]:
    pd.fill(c)
    pd.display()
    print('fill:', n)
    delay(500)

pd.fill(BLACK)
pd.display()
for y in range(pd.height):
    for x in range(pd.width):
        pd.setColor(x, y, WHITE)
        pd.display()
        delay(20)

delay(300)
pd.fill(BLACK)
pd.display()
print('Display test done')
````

````python
import subprocess

def run(cmd):
    print('$', cmd)
    try:
        out = subprocess.check_output(cmd, shell=True, stderr=subprocess.STDOUT, text=True)
        print(out.strip() if out.strip() else '(no output)')
    except subprocess.CalledProcessError as e:
        print(e.output.strip() if e.output else str(e))
    print('-' * 60)

for c in [
    'id',
    'groups',
    'uname -a',
    'ls -l /dev/mem',
    'ls -l /dev/gpiomem',
    'ls -l /dev/vcio'
]:
    run(c)
````

## Terminal Cross-check (permission difference)
Run these in Jupyter Terminal (not Python cell):

```bash
python3 - <<'PY'
from pop import PixelDisplay
pd = PixelDisplay(automode=False)
pd.fill(pd.WHITE if hasattr(pd, 'WHITE') else 0xFFFFFF)
pd.display()
print('non-root test done')
PY

sudo python3 - <<'PY'
from pop import PixelDisplay
pd = PixelDisplay(automode=False)
pd.fill(pd.WHITE if hasattr(pd, 'WHITE') else 0xFFFFFF)
pd.display()
print('root test done')
PY
```
If root works but non-root does not, it is likely a permission/access issue.
