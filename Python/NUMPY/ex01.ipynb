```python
import numpy as np

angles = np.linspace(0, 180, 11)      # 0~180을 11개로 균등 분할
sin_values = np.sin(np.deg2rad(angles))  # sin은 라디안 입력이므로 변환
list_sin = []
for a, s in zip(angles, sin_values):
    print(f"sin({int(a):3d}°) = {s:.6f}")
    list_sin.append(s)
    
print(list_sin)
```
